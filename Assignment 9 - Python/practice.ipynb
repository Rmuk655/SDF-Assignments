```python
# Pig Latin (http://mng.bz/YrON) is a common children’s secret language in Englishspeaking countries. The rules for translating words from English into
# Pig Latin are quite simple: If the word begins with a vowel (a, e, i, o, or u), add “way” to the end of the word. So “air” becomes “airway” and
# “eat” becomes “eatway”. If the word begins with any other letter, then we take the first letter, put it on the end of the word, and then add “ay”. Thus,
# “python” becomes “ythonpay” and “computer” becomes “omputercay”. Convert the following sentence “this is a test” to a Pig-Latin.
def piglatin(s):
    if(s[0] == "a" or s[0] == "e" or s[0] == "i" or s[0] == "o" or s[0] == "u"):
        print(s + "way")
    else:
        print(s[1:] + s[0] + "ay")
s = "this is a test"
x = s.split()
for i in x:
    piglatin(i)
```

```python
# Merging Overlapping Intervals: You are given a list of time intervals as (start, end) tuples.
# Your task is to merge overlapping intervals into a single interval whenever possible.
def merge(m):
    s = []
    for i in range(len(m) - 1):
        if(m[i][1] > m[i + 1][0]):
            t = (m[i][0], m[i + 1][1])
            s.append(t)
    if(s == []):
        s = m
    print(s)
intervals1 = [(1, 3), (2, 6), (8, 10), (9, 11)]
intervals2 = [(1, 5), (6, 10), (11, 15)]
merge(intervals1)
merge(intervals2)
```

```python
# Create a new constant dict, called MENU, representing the possible items you can order at a restaurant. The keys will be strings, and the values will be prices 
# (i.e., integers). You should then write a function, restaurant, that asks the user to enter an order: If the user enters the name of a dish on the menu, the program prints 
# the price and the running total. It then asks the user again for their order. If the user enters the name of a dish not on the menu, the program scolds the user (mildly). 
# It then asks the user again for their order. If the user enters an empty string, the program stops prompting and prints the total amount. 
# You can order the same order multiple times. Inputs should be case-insensitive.
MENU = {'sandwich': 10, 'tea': 7, 'salad': 9}
def restaurant():
    x = input("Enter the item you want to order")
    x = x.lower()
    r = 0
    while(x != ""):
        if (x in MENU):
            print("Price = ", MENU[x])
            r = r + MENU[x]
            print("Running Total = ", r)
        else:
            print("Enter an item in the order.")
        x = input("Enter the item you want to order")
        x = x.lower()
restaurant()
```

```python
# Write a function, get_rainfall, that tracks rainfall in a number of cities. Users of your program will enter the name of a city; if the city name is blank, then the function
# prints a report (city_name: rainfall amount) before exiting. If the city name isn’t blank, then the program should also ask the user how much rain has fallen in that
# city (typically measured in millimeters). After the user enters the quantity of rain, the program again asks them for a city name, rainfall amount, and so on—until the user
# presses Enter instead of typing the name of a city. When the user enters a blank city name, the program exits—but first, it reports how much total rainfall there was in
# each city. The order in which the cities appear is not important, and the cities aren’t known to the program in advance.
def get_rainfall():
    x = input("Enter the city name: ")
    d = {}
    while (x != ""):
        y = int(input("Enter the amount of rainfall: "))
        d[x] = y
        x = input("Enter the city name: ")
    if (x == ""):
        for i in d:
            print(i, ":", d[i])
get_rainfall()
```

```python
# Write a function, myxml, that allows you to create simple XML output. The output from the function will always be a string. The function can be invoked in a number of ways, 
# as shown in the table. Input: 'tagname', 'hello', a=1, b=2, c=3)
def myxml(tag, content, **attrs):
    attributes = "".join(f' {key}="{value}"' for key, value in attrs.items())
    print(f"<{tag}{attributes}>{content}</{tag}>") 
myxml('tagname', 'hello', a=1, b=2, c=3)
```

```python
# Write a function (calc) that expects a single argument—a string containing a simple math expression in prefix notation—with an operator and two numbers.
# Your program will parse the input and produce the appropriate output. For our purposes, it’s enough to handle the six basic arithmetic operations in Python:
# addition (+), subtraction (-), multiplication (*), division (/), modulus (%), and exponentiation (**). The normal Python math rules should work, such that division 
# always results in a floating-point number. We’ll assume, for our purposes, that the argument will only contain one of our six operators and two valid numbers.
# Input: + 2 3
def calc(s):
    x = s.split()
    print(x)
    if (x[0] == "+"):
        return (int(x[1]) + int(x[2]))
    elif (x[0] == "-"):
        return (int(x[1]) - int(x[2]))
    elif (x[0] == "*"):
        return (int(x[1]) * int(x[2]))
    elif (x[0] == "/"):
        return (float(x[1])/float(x[2]))
    elif (x[0] == "%"):
        return (int(x[1]) % int(x[2]))
    elif (x[0] == "**"):
        return (int(x[1])**int(x[2]))
s = "+ 2 3"
print(calc(s))
```

```python
# Write a program to evaluate any general prefix expression.
import operator
def pcalc(s):
    ops = {"+": operator.add, "-": operator.sub, "*": operator.mul, "/": operator.truediv}
    stack = []
    tokens = s.split()
    for token in reversed(tokens):
        if token.isdigit():  # If token is a number, push it to the stack
            stack.append(int(token))
        elif token in ops:  # If token is an operator, apply it to top two elements in stack
            a = stack.pop()
            b = stack.pop()
            result = ops[token](a, b)  # Perform operation
            stack.append(result)  # Push result back to stack

    return stack[0] 
s = "* 2 + 3 4"
print(pcalc(s))
```

```python
# OOPS problem
# c_q = customer_queue, a_a = agent_availability
import time

class CallCenter:
    def __init__(self):
        self.c_q = []  # Customer Queue
        self.agent_available = True  # Initially available

    def add_customer(self, name, t):
        self.c_q.append((name, t))  # Store customer with their required time
        print(f"{name} added to the queue.")

    def serve_customer(self):
        if not self.c_q:
            print("no customer")
            return

        if not self.agent_available:
            print("Sorry, currently serving another customer.")
            return

        customer, service_time = self.c_q.pop(0)
        self.agent_available = False
        print(f"Serving customer: {customer} for {service_time} seconds...")

        time.sleep(service_time)  # Simulate service time

        self.agent_available = True
        print(f"Finished serving {customer}. Agent is now available.")

    def show_queue(self):
        if not self.c_q:
            print("empty queue")
        else:
            print("Current queue:", [customer[0] for customer in self.c_q])  # Show only names


# ---------------- Test Run ----------------
call_center = CallCenter()

while True:
    print("\nOptions: 1. Add Customer  2. Serve Customer  3. Show Queue  4. Exit")
    choice = input("Enter choice: ")

    if choice == "1":
        name = input("Enter customer name: ")
        service_time = int(input("Enter service time (seconds): "))
        call_center.add_customer(name, service_time)

    elif choice == "2":
        call_center.serve_customer()

    elif choice == "3":
        call_center.show_queue()

    elif choice == "4":
        print("Exiting Call Center simulation.")
        break

    else:
        print("Invalid choice! Please try again.")
```

```python
class Coordinate:
    def __init__(self, x, y):
        self.x = x
        self.y = y
    def distance(self, other):
        a = (self.x - other.x)**2
        b = (self.y - other.y)**2
        c = (a + b)**0.5
        return c
c = Coordinate(2, 3)
d = Coordinate(3, 4)
print(Coordinate.distance(c, d))
print(c)
```
